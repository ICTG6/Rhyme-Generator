**WEB SCRAPING**

1. Import Libraries

```
from bs4 import BeautifulSoup # html parser
import requests # request html link
import csv
```

2. Get URL webpage

```
# get url and request access 
url = "https://www.rhymedb.com/word/air" 
page = requests.get(url)
pagetext = page.text
soup = BeautifulSoup (pagetext,'html.parser') 
```

3. Scrape the webpage

```
# find keywords to identify words
word_rhyme = soup.findAll('a', {'id':'t2'}) 
# sort element in alphabetical order
word_rhyme = sorted(word_rhyme, key=lambda elem: elem.text)
```

4. Write to CSV FILE

```
# create new file and write into csv file
filename ='air_rhyme.csv' 
f = open(filename,'w')
headers= "Word_rhyme_with_air\n" #[table header words]
f.write(headers)
```

5. Get all words and write to csv File

```
# going through a loop based on number of words that is included
for word in word_rhyme:
   word = str(word) 
   
   word=word[33:60]
    # use the if statement because some oputput give extra word, replace it with empty'
   if "word/" in word:
      word = word.replace("word/",'')
      word= word.replace('"','') # change all that contain "" into''(empty), erase it
   elif "id=t2>" in word:
       word = word.replace("id=t2>",'')
   elif "d=t2>" in word:
       word = word.replace("d=t2>",'')
   else:
    word= word.replace('"','')
   # only take the word in index[0] after split
   y = (word.split(' ')[0]) 

   # write to the csv file
   f.write(y+"\n")

   # print all the word that is written to the csv file
   print(y)
```
